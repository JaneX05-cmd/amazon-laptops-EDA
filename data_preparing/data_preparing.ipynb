```python
import csv
import io

def clean_without_last_col(input_file, output_file):
    header_names = [
        'id', 'product_title', 'rating', 'total_rating', 'price', 
        'next_month_bought', 'offer', 'mrp', 'product_details_dict', 
        'brand', 'model_name', 'screen_size', 'colour', 'hard_disk_size', 
        'cpu_model', 'ram', 'os', 'special_features', 'graphics_card', 
        'about'
    ]
    
    with open(input_file, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    clean_rows = []
    for line in lines[1:]:
        raw_line = line.strip().strip('"').rstrip(';')
        raw_line = raw_line.replace('""', '"')
        
        reader = csv.reader(io.StringIO(raw_line))
        try:
            fields = next(reader)
        except StopIteration:
            continue
            
        clean_rows.append(fields[:20])

    with open(output_file, 'w', encoding='utf-8', newline='') as f:
        writer = csv.writer(f, quoting=csv.QUOTE_MINIMAL)
        writer.writerow(header_names)
        writer.writerows(clean_rows)

clean_without_last_col('/Users/janek/Downloads/single_amazon_laptop_messy_dataset_1-10.csv', '/Users/janek/Documents/Data Analysis Learning/EDA Project/csv_laptop_clean.csv')
```
